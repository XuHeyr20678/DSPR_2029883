```python
bin_posx = [-0.5260,-0.3794,-0.1448,-0.1371, -0.1367,-0.0884,-0.0496, -0.0495, -0.0312, -0.0310,-0.0069,0.0095,0.0096, 0.0108,0.0109, 0.0324,0.0332, 0.0333,0.0566]
bin_posy = [-0.0077,-0.0011,-0.000506,-0.000504, -0.000079,-0.000025, 0.0649, 0.1728,0.4660,1.0987,1.3385,1.3782,1.4326,1.5042]
bin_vlex = [-0.8075,-0.7203,-0.6523, -0.6260, -0.4819,-0.1684,-0.0044,-0.0001,0.000097,0.0057,0.1777,0.3404,0.5080]
bin_vely = [-0.6844,-0.3679, -0.1831,-0.0059,-0.0004,-0.000001,0,0.0004,0.0771, 0.2109,0.4549]
bin_ang =[-0.6011,-0.4160,-0.1136,-0.0056, 0.0003,0.0038,0.0078,0.0273,0.2488]
bin_angvel = [-0.3034,-0.1671,-0.0796,-0.0003,-0.000092,0.000045,0.0010,0.0448,0.1489,0.3607]
```

```python
pos_x = pd.cut(df['pos x'], bin_posx).astype(str)
pos_y = pd.cut(df['pos y'], bin_posy).astype(str)
vel_x = pd.cut(df['vel x'], bin_vlex).astype(str)
vel_y = pd.cut(df['vel y'], bin_vely).astype(str)
angle = pd.cut(df['angle'], bin_ang).astype(str)
angleVel = pd.cut(df['angleVel'], bin_angvel).astype(str)
contact_1 =pd.cut(df['contact_1'], 2).astype(str)
contact_2 =pd.cut(df['contact_2'], 2).astype(str)
df['pos x'] = pos_x
df['pos y'] = pos_y
df['vel x'] = vel_x
df['vel y'] = vel_y
df['angle'] = angle
df['angleVel'] = angleVel
df['contact_1'] = contact_1
df['contact_2'] = contact_2
```

```python
df
```

```python
def getlevelx(a):
    if a==0:
        return "No action"
    elif a==1:
        return "Fire left engine"
    elif a==2:
        return "Fire down engine"
    else:
        return "Fire right engine"
```

```python
df['class']=df.apply(lambda x: getlevelx(x['action']),axis=1)
df
```

```python
df_new = df.drop(columns=['action'])
df_new
```

```python
train, test = train_test_split(df_new, test_size=0.3)
train.to_csv('train_MDLP.csv', index=False)
test.to_csv('test_MDLP.csv', index=False)
```
