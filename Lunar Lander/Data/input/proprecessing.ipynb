```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('original_data.csv')
df
```

```python
pos_x = pd.cut(df['pos x'], 6).astype(str)
pos_y = pd.cut(df['pos y'], 6).astype(str)
vel_x = pd.cut(df['vel x'], 6).astype(str)
vel_y = pd.cut(df['vel y'], 6).astype(str)
angle = pd.cut(df['angle'], 6).astype(str)
angleVel = pd.cut(df['angleVel'], 6).astype(str)
df['pos x'] = pos_x
df['pos y'] = pos_y
df['vel x'] = vel_x
df['vel y'] = vel_y
df['angle'] = angle
df['angleVel'] = angleVel
```

```python
df
```

```python
df.dtypes
```

```python
contact_1 = pd.cut(df['contact_1'],2).astype(str)
contact_2 = pd.cut(df['contact_2'],2).astype(str)
df['contact_1']=contact_1
df['contact_2']=contact_2
```

```python
def getlevelx(a):
    if a==0:
        return "No action"
    elif a==1:
        return "Fire left engine"
    elif a==2:
        return "Fire down engine"
    else:
        return "Fire right engine"
```

```python
df['class']=df.apply(lambda x: getlevelx(x['action']),axis=1)
df
```

```python
df_new = df.drop(columns=['action'])
df_new
```

```python
df_new.dtypes
```

```python
train, test = train_test_split(df_new, test_size=0.3)
train.to_csv('train_1.csv', index=False)
test.to_csv('test_1.csv', index=False)
```
